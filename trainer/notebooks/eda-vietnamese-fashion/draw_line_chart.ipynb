```python
import matplotlib.pyplot as plt

# Data
methods = ['HashNet', 'GreedyHash', 'CSQ', 'OrthoHash', 'RelaHash']
x_values = [16, 128, 2048]
y_values = [
    [0.2384, 0.3742, 0.3865],
    [0.2495, 0.3835, 0.377],
    [0.1945, 0.3379, 0.2948],
    [0.276, 0.4332, 0.4741],
    [0.2967, 0.4546, 0.4954],
]

# Create the line chart (with a small size just for extracting the legend)
plt.figure(figsize=(2, 2))

for i, method in enumerate(methods):
    plt.plot(x_values, y_values[i], marker='o', label=method)

plt.legend(fontsize=20)
plt.axis('off')  # Turn off axis and ticks
plt.grid(False)  # Hide the grid

# hide lines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


# Save only the legend with transparent background
plt.savefig('legend_transparent.png', transparent=True, bbox_inches='tight', pad_inches=0)

plt.show()
```
